```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#load the data and store the data in data frames
df_btc = pd.read_csv("Bitcoin Historical Data - Investing.com India.csv")
df_eth = pd.read_csv("Ethereum Historical Data - Investing.com India.csv")
df_bnb = pd.read_csv("Binance Coin Historical Data - Investing.com India.csv")
```

```python
#print the data
print(df_btc)
print(df_eth)
print(df_bnb)
```

```python
#create a new dataframe that holds the all 3 crypto price(closing)
df = pd.DataFrame({'BTC':df_btc['Price'],
                   'ETH':df_eth['Price'],
                   'BNB':df_bnb['Price']
})
```

```python
#show new dataframe
df
```

```python
#stats of the data
df.describe()
```

```python
#visualzing the price(closing)
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

```

```python
my_crypto = df
plt.figure(figsize=(12.2,4.5))
for c in my_crypto.columns.values:
    plt.plot(my_crypto[c], label =c)

plt.title("Latest Crypto Graph")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend(my_crypto.columns.values, loc= 'upper left')
plt.show()
```

```python
#scale the data
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0, 100))
scaled = min_max_scaler.fit_transform(df)
scaled
```

```python
#Convert the scaled data into a dataframe
df_scale = pd.DataFrame(scaled, columns = df.columns)
```

```python
#visualize the scaled data
my_crypto = df_scale
plt.figure(figsize=(12.5, 4.5))
for c in my_crypto.columns.values:
    plt.plot(my_crypto[c], label=c)

plt.title('Crypto Scaled Graph')
plt.xlabel('Days')
plt.ylabel('Price')
plt.legend(my_crypto.columns.values, loc = "upper left")
plt.show()
```

```python
#Get the daily simple return
DSR = df.pct_change(1)
DSR
```

```python
#Visualize the daily simple returns
plt.figure(figsize=(12,4.5))
for c in DSR.columns.values:
    plt.plot(DSR.index, DSR[c], label = c, lw = 2,alpha = .7)

plt.title('Daily Simple Returns')
plt.ylabel('Percentage (in decimal form')
plt.xlabel('Days')
plt.legend(DSR.columns.values, loc ='upper right')
plt.show()
```

```python
#Get the volatility:
print("The crypto volatility:")
DSR.std()
```

```python
#show the mean/average daily simole return
DSR.mean()
```

```python
#Get the correlation
DSR.corr()
```

```python
#visualize the correlation
import seaborn as sns
plt.subplots(figsize=(11,11))
sns.heatmap(DSR.corr(), annot=True, fmt='.2%')
```

```python
#Get the daily cumulative simple returns
DCSR = (DSR+1).cumprod()
DCSR
```

```python
# visualize the daily cumulative simple returns
plt.figure(figsize=(12.2,4.5))
for c in DCSR.columns.values:
    plt.plot(DCSR.index, DCSR[c], label = c, lw = 2,alpha = .7)

plt.title('Daily Cumulative Simple Returns')
plt.xlabel('Days')
plt.ylabel('Growth of 1$ investment')
plt.legend(DCSR.columns.values, loc ='upper left',fontsize=10)
plt.show()
```
